```python
#Theory Questions

"""
#Question 1
What is Python, and why is it popular?
-Python is a high-level, easy-to-read programming language used for everything from web development to AI and data science. It’s popular because its syntax is simple, beginner-friendly, and highly productive. Python also has massive libraries and community support, making it powerful for both small scripts and large applications.

#Question 2
What is an interpreter in Python?
-An interpreter executes Python code line by line, translating it into machine instructions at runtime. This makes debugging easier and allows Python code to run on different platforms without recompilation. It’s one of the main reasons Python is flexible and easy to test.

#Question 3
What are pre-defined keywords in Python?
-Keywords are reserved words that Python uses for its syntax and structure. They have fixed meanings and cannot be redefined or used as variable/function names. Examples include: if, else, while, class, import, True, False.

#Question 4
Can keywords be used as variable names?
-No, keywords cannot be used as variable names. Because they control Python’s syntax, using them as identifiers would create ambiguity and errors.

#Question 5
What is mutability in Python?
-Mutability is the ability of an object to change its value after creation. Mutable objects can be updated in place, while immutable objects create a new copy when modified.

#Question 6
Why are lists mutable, but tuples are immutable?
-Lists are mutable because they are designed for dynamic data, where elements may need to be added, changed, or removed.
-Tuples are immutable to provide fixed, reliable, and faster data structures—useful when data should not change, like coordinates or constants.

#Question 7
What is the difference between “==” and “is” operators in Python?
- == checks value equality—whether two objects contain the same data.
- is checks identity—whether both references point to the exact same object in memory.
  Thus, two objects can be == but not is.

#Question 8
What are logical operators in Python?
-Logical operators are used to combine conditions and return a Boolean result.
  They include:

  and → true if both conditions are true

  or → true if any one condition is true

  not → reverses the truth value

#Question 9
What is type casting in Python?
-Type casting means converting one data type into another. Python provides built-in functions like int(), float(), str(), list(), etc. It helps ensure compatibility between values in expressions.

#Question 10
What is the difference between implicit and explicit type casting?
-Implicit casting is done automatically by Python when needed (e.g., int → float during arithmetic).
-Explicit casting is done manually by the programmer using functions like int(), float(), str().
  Explicit casting gives more control and prevents unexpected errors.

#Question 11
What is the purpose of conditional statements in Python?
-Conditional statements allow a program to make decisions and execute code only when certain conditions are satisfied. They help control the flow of execution using if, elif, and else. This is essential for logic-based programming.

#Question 12
How does the elif statement work?
-elif allows checking multiple conditions sequentially. Python evaluates conditions from top to bottom and executes the first one that becomes true. If none are true, the else block (if present) runs.

#Question 13
What is the difference between for and while loops?
-for loop: Used when you know the number of iterations or when iterating over a sequence (list, range, string).

-while loop: Runs until a condition becomes false, useful when the number of iterations is not fixed.
  for is structured; while is condition-driven.

#Question 14
Describe a scenario where a while loop is more suitable than a for loop.
-A while loop is better when the program depends on a changing condition, not fixed repetitions.
  Example: Keep asking a user to enter a correct PIN until they enter the right one. Here, you don’t know how many attempts the user will take.

"""
```

```python
#Coding Questions

##Question 1
#Write a Python program to print "Hello, World!
print("Hello, World!")


##Question 2
#Write a Python program that displays your name and age
name = input("Enter your name: ")
age = input("Enter your age: ")
print("Name:", name)
print("Age:", age)


##Question 3
#Write code to print all the pre-defined keywords in Python using the keyword library
import keyword
print("Python Keywords:")
print(keyword.kwlist)


##Question 4
#Write a program that checks if a given word is a Python keyword.
import keyword

word = input("Enter a word: ")
if keyword.iskeyword(word):
    print(word, "is a Python keyword.")
else:
    print(word, "is NOT a Python keyword.")


##Question 5
#Write a function to demonstrate the behavior of mutable and immutable arguments.
def change_list(a):
    a.append(100)
    print("Inside function (list):", a)

def change_number(n):
    n = n + 10
    print("Inside function (number):", n)

my_list = [1, 2, 3]
my_num = 5

print("Before:", my_list, my_num)
change_list(my_list)
change_number(my_num)
print("After: ", my_list, my_num)


##Question 6
#Write a program that performs basic arithmetic operations on two user-input numbers.
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)

if b != 0:
    print("Division:", a / b)
else:
    print("Cannot divide by zero.")


##Question 7
#Write a program to demonstrate the use of logical operators.
x = int(input("Enter a number: "))

print("x > 0 and x < 10:", x > 0 and x < 10)
print("x < 0 or x > 10:", x < 0 or x > 10)
print("not (x == 5):", not (x == 5))


##Question 8
#Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.
my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

print("Original list:", my_list)
my_list[1] = 99
print("Modified list:", my_list)

print("Original tuple:", my_tuple)
try:
    my_tuple[1] = 99
except TypeError:
    print("Error: Cannot change elements of a tuple (immutable)")


##Question 9
#Write a Python program to convert user input from string to integer, float, and boolean types.
value = input("Enter something: ")

try:
    print("As integer:", int(value))
except:
    print("Cannot convert to integer.")

try:
    print("As float:", float(value))
except:
    print("Cannot convert to float.")

# simple boolean conversion
if value.lower() in ["true", "1", "yes"]:
    print("As boolean: True")
elif value.lower() in ["false", "0", "no"]:
    print("As boolean: False")
else:
    print("Cannot convert to boolean.")


##Question 10
#Write code to demonstrate type casting with list elements.
lst = ["10", "20", "30"]
print("Original list:", lst)

int_list = [int(i) for i in lst]
float_list = [float(i) for i in lst]

print("As integers:", int_list)
print("As floats:", float_list)


##Question 11
#Write a program that checks if a number is positive, negative, or zero.
num = float(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")


##Question 12
#Write a for loop to print numbers from 1 to 10.
for i in range(1, 11):
    print(i)


##Question 13
#Write a Python program to find the sum of all even numbers between 1 and 50.
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i

print("Sum of even numbers from 1 to 50:", total)


##Question 14
#Write a program to reverse a string using a while loop.
s = input("Enter a string: ")
i = len(s) - 1
rev = ""

while i >= 0:
    rev += s[i]
    i -= 1

print("Reversed string:", rev)


##Question 15
#Write a Python program to calculate the factorial of a number provided by the user using a while loop.
n = int(input("Enter a number: "))

fact = 1
i = 1

while i <= n:
    fact *= i
    i += 1

print("Factorial of", n, "is", fact)
```
